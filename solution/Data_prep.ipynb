```python
import pandas as pd
import pytesseract
from PIL import Image
import requests
from io import BytesIO
from concurrent.futures import ThreadPoolExecutor
from tqdm import tqdm


config = r'--oem 2'


def process_image(url):
    try:
        # Download the image
        response = requests.get(url)
        image = Image.open(BytesIO(response.content))

        # Perform OCR
        # text = pytesseract.image_to_string(image,conifg=config)
        text = pytesseract.image_to_string(image)

        return text.strip()
    except Exception as e:
        print(f"Error processing {url}: {str(e)}")
        return None

def process_dataset(df, num_workers=10):
    with ThreadPoolExecutor(max_workers=num_workers) as executor:
        # Create a TQDM progress bar
        results = list(tqdm(
            executor.map(process_image, df['image_link']),
            total=len(df),
            desc="Processing Images"
        ))

    df['extracted_text'] = results
    return df
```

```python
# Process the dataset
processed_df = process_dataset(df[0:8])

# Save the results
processed_df.to_csv('processed_dataset.csv', index=False)
```
